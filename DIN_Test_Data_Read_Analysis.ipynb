```python
#calculating total run time 
Total_run_episode=[]
#Rl_runTime=[]
#SA_runTime=[]
for key in df.keys():
    rl=df[key]["rl_duration"]
    sa=df[key]["sa_duration"]
    Total_run_episode.append(rl+sa)
plt.plot(Total_run_episode) #why there is an increasing trend this is not cumulative
plt.ylabel("Runtime (cpu seconds)")
plt.xlabel("Episode Number")
plt.title("Runtime per episode")
plt.grid()
plt.savefig('runtimeperepisode.pdf') 
```

```python
#lets plot cumulative 
plt.plot (np.cumsum(Total_run_episode)) #ok almost linear but there is expnential trend here
#I might put box plot next to this for each episode 
plt.ylabel("Runtime (cpu seconds)")
plt.xlabel("Episode Number")
plt.title("Cummulative Runtime after each Episode")
plt.grid()
plt.savefig('cumulativeruntime.pdf') 
```

```python
#Convergence of best and current solution 
best_solution_stage=[] 
rl_best=[]
sa_best=[]
for key in df.keys():
    rl_best.append(df[key]["rl"][-1]["best_rl_cost"]) #this is the best best objective value found at the end of rl stage
    sa_best.append(df[key]["sa"][-1]["best_sa_cost"]) #this is the best best objective value found at the end of sa stage
    
    
```

```python
plt.plot (rl_best, label='DRL convergence')
plt.plot (sa_best, label='SA convergence')
plt.xlabel("Episode Number")
plt.title("Best Objective Function Value Found by RL and SA")
plt.ylabel(r"Objective Function Value $(\theta=1)$ ")
plt.grid()
plt.legend()

plt.savefig('convergence.pdf') 
```

```python
#First episode in more detail plotting 
rl_current=[]
rl_best=[]
sa_current=[]
sa_best=[]
for i in df["1"]["rl"]:
    rl_best.append(i["best_rl_cost"])
    rl_current.append(i['state_cost'])
        
for i in df["1"]["sa"]:
    sa_best.append(i["best_sa_cost"])
    sa_current.append(i['state_cost'])

sa_best=[rl_best[-1]]+sa_best
#I need to add initial sa as SA best at iteration 0
x=range(len(rl_best)+len(sa_best))
#print(len(sa_best))


plt.figure(figsize=(12,7))

plt.plot (x[:len(rl_best)],rl_best, label='the best solution found by RL')
plt.plot (x[len(rl_best):],sa_best, label='the best solution found by SA')

plt.plot(rl_current+[rl_current[-1]]+sa_current, label="current solution", linestyle='dashed', linewidth=0.7)

plt.legend()

plt.xlabel("Time step/SA iteration")
plt.ylabel(r"Objective Function Value $(\theta=1)$ ")
plt.grid()
plt.title("Objective function value progress in a single episode")
plt.savefig('convergenceperepisode.pdf') 

#plt.plot (rl_current+sa_current)
```
